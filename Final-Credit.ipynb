## Data Cleaning on applicaton_dataset

```python
# Describing each numberical column of application_dataset
df_app.describe()
```

```python
# Listing all the columns present in the application_dataset
df_app.columns
```

```python
# Checking the null values 
df_app_null = df_app.isnull().sum()
df_app_null
```

```python
# Calculating the null values percentage

df_app_null_percent = round(100*df_app_null/len(df_app), 2)
df_app_null_percent
```

```python
# Total number of columns having null values
null_cols = df_app_null[df_app_null != 0].count()
print(null_cols)
```

```python
# Calculates the mean (average) null value per column in the dataframe
# Same with above calculating but simple code
df_app.isnull().mean()
```

```python
# Our team decide to removing null values more than 19%
df_app = df_app.loc[:, df_app.isnull().mean() <= .19]
```

```python
# Number of columns left after removal of the null values < 19%
len(df_app.columns)
```

```python
# Checking the data again
round(100*df_app.isnull().sum()/len(df_app), 2)
```

## Inputing the missing values

### Inputing the missing values in 'application_data'

### Categorical columns

#### 1. NAME_TYPE_SUITE: Who was accompanying client when he was applying for the loan

```python
df_app.NAME_TYPE_SUITE.value_counts()
```

```python
# Checking the mode value in the NAME_TYPE_SUITE column
df_app.NAME_TYPE_SUITE.mode()
```

```python
# Filling the 'NAME_TYPE_SUITE' columns with the mode value 'Unaccompanied'
# as it the most occurring value to the column
df_app['NAME_TYPE_SUITE'].fillna(value = 'Unaccompanied', inplace=True)
```

### Numerical columns
