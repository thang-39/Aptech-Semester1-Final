```python
df_app.NAME_TYPE_SUITE.value_counts()
```

```python
# Checking the mode value in the NAME_TYPE_SUITE column
df_app.NAME_TYPE_SUITE.mode()
```

```python
# Filling the 'NAME_TYPE_SUITE' columns with the mode value 'Unaccompanied'
# as it the most occurring value to the column
df_app['NAME_TYPE_SUITE'].fillna(value = 'Unaccompanied', inplace=True)
```

```python
# in this case, our team assume that null will be 0
```

```python
df_app.loc[:, 'AMT_REQ_CREDIT_BUREAU_HOUR':'AMT_REQ_CREDIT_BUREAU_YEAR'] = df_app.loc[:, 'AMT_REQ_CREDIT_BUREAU_HOUR':'AMT_REQ_CREDIT_BUREAU_YEAR'].fillna(0)
```

#### 3. AMT_GOODS_PRICE column: For consumer loans it is the price of the goods for which the loan is given

```python
df_app.AMT_GOODS_PRICE.describe()
```

```python
from sklearn.impute import KNNImputer

# Create KNNImputer instance
imputer = KNNImputer(n_neighbors=3)
```

```python
# Extract the column with null values
dataset_column = df_app['AMT_GOODS_PRICE'].values.reshape(-1, 1)


# Fit the imputer on the column
imputer.fit(dataset_column)

# Transform the column to replace null values
imputed_column = imputer.transform(dataset_column)

# Assign the imputed values back to the dataset
df_app['AMT_GOODS_PRICE'] = imputed_column
```

```python
df_app.AMT_GOODS_PRICE.describe()
```

```python
# check null of data_app
df_app.isnull().mean()
```

#### 4. AMT_ANNUITY column: Loan annuity: refers to the fixed amount of money that a borrower is required to pay periodically to repay a loan

```python
df_app.AMT_ANNUITY.value_counts() 
```

```python
df_app[df_app['AMT_ANNUITY'].isna()]
```

```python
# Extract the column with null values
dataset_column = df_app['AMT_ANNUITY'].values.reshape(-1, 1)

# Fit the imputer on the column
imputer.fit(dataset_column)

# Transform the column to replace null values
imputed_column = imputer.transform(dataset_column)

# Assign the imputed values back to the dataset
df_app['AMT_ANNUITY'] = imputed_column
```

```python
# check null of data_app
df_app.isnull().mean()
```
