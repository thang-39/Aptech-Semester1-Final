# StandardScaler

```python
# Bien doi cac thuoc tinh co phan phoi Gaussian, mean, do lech chuan (std) khac nhau thanh phan phoi Guassian tieu chuan 
# voi gia tri trung binh la 0 va do lech chuan la 1
# (xi - mean(x)) /std(x)
# linear regression, logistic regression
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
np.random.seed(1)
df = pd.DataFrame({
    'A': np.random.normal(0,2,10000),
    'B': np.random.normal(5,3,10000),
    'C': np.random.normal(-5,5,10000),
    
})
```

```python
df.head()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
sns.distplot(df['A'])
plt.subplot(1,3,2)
sns.distplot(df['B'])
plt.subplot(1,3,3)
sns.distplot(df['C'])
plt.show()
```

```python
print('A mean: ', df['A'].mean())
print('B mean: ', df['B'].mean())
print('C mean: ', df['C'].mean())
```

```python
print('A std: ', df['A'].std())
print('B std: ', df['B'].std())
print('C std: ', df['C'].std())
```

```python
f, ax1 = plt.subplots(ncols=1, figsize=(6,5))
sns.distplot(df['A'], ax=ax1, hist=False)
sns.distplot(df['B'], ax=ax1, hist=False)
sns.distplot(df['C'], ax=ax1, hist=False)
plt.show()
```

```python
f, ax1 = plt.subplots(ncols=1, figsize=(6,5))
sns.kdeplot(df['A'], ax=ax1)
sns.kdeplot(df['B'], ax=ax1)
sns.kdeplot(df['C'], ax=ax1)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df_scale = scaler.fit_transform(df)
df_scale = pd.DataFrame(df_scale, columns=['A','B','C'])
```

```python
df_scale.head()
```

```python
print('A mean after scale: ', df_scale['A'].mean())
print('B mean after scale: ', df_scale['B'].mean())
print('C mean after scale: ', df_scale['C'].mean())
```

```python
print('A std after scale: ', df_scale['A'].std())
print('B std after scale: ', df_scale['B'].std())
print('C std after scale: ', df_scale['C'].std())
```

```python
f, (ax1,ax2) = plt.subplots(ncols=2, figsize=(8,5))
sns.kdeplot(df['A'], ax=ax1)
sns.kdeplot(df['B'], ax=ax1)
sns.kdeplot(df['C'], ax=ax1)
sns.kdeplot(df_scale['A'], ax=ax2)
sns.kdeplot(df_scale['B'], ax=ax2)
sns.kdeplot(df_scale['C'], ax=ax2)
plt.show()
```

```python
df.corr()
```

```python
df_scale.corr()
```

```python
# MinMaxScaler
# (xi - min(x))/ (max(x) - min(x))
# thu hep pham vi sao pham vi moi nam trong khoang tu 0 den 1 ( hoac la -1 den 1 neu co cac gia tri am).
# Neu phan phoi khong la phai Gaussian hoac de lech chuan rat nho, min-max scaler hoat dong tot
# Rat nhay voi outlier - > ap dung Robust Scaler
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('Data_for_Values.csv')
df
```

```python
df_new = df[['Age', 'Salary']]
```

```python
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
sns.distplot(df_new['Age'])

plt.subplot(1,2,2)
sns.distplot(df_new['Salary'])
plt.show()
```

```python
df_new.Age.skew(), df_new.Salary.skew()
```

```python
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.boxplot(df_new['Age'])

plt.subplot(1,2,2)
plt.boxplot(df_new['Salary'])
plt.show()
```

```python
# => khng co outlier , khng phi phn phi chun => dng MinMaxScaler
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df_scale = scaler.fit_transform(df_new)
df_scale = pd.DataFrame(df_scale, columns=['Age', 'Salary'])
```

```python
df_scale
```

```python
print('Age min before scale: ', df_new['Age'].min())
print('Age max before scale: ', df_new['Age'].max())
print('Salary min before scale: ', df_new['Salary'].min())
print('Age max before scale: ', df_new['Salary'].max())
```

```python
print('Age min after scale: ', df_scale['Age'].min())
print('Age max after scale: ', df_scale['Age'].max())
print('Salary min after scale: ', df_scale['Salary'].min())
print('Age max after scale: ', df_scale['Salary'].max())
```

```python
f, (ax1,ax2) = plt.subplots(ncols=2, figsize=(8,5))

ax1.set_title("Before Scaling")
sns.kdeplot(df_new['Age'], ax=ax1)
sns.kdeplot(df_new['Salary'], ax=ax1)

ax2.set_title("After Scaling")
sns.kdeplot(df_scale['Age'], ax=ax2)
sns.kdeplot(df_scale['Salary'], ax=ax2)

plt.show()
```

```python
df.corr()
```

```python
df_scale.corr()
```

```python
#Robust Scaler: 
# - tng t MinMax nhng n s dng Interquartile range thay cho min-max
# - cng thc:(xi - Q1(x))/(Q3(x) - Q1(x))
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
```

```python
df_scale = scaler.fit_transform(df_new)
df_scale = pd.DataFrame(df_scale, columns=['Age', 'Salary'])
df_scale
```

```python
f, (ax1,ax2) = plt.subplots(ncols=2, figsize=(8,5))

ax1.set_title("Before Scaling")
sns.kdeplot(df_new['Age'], ax=ax1)
sns.kdeplot(df_new['Salary'], ax=ax1)

ax2.set_title("After Scaling")
sns.kdeplot(df_scale['Age'], ax=ax2)
sns.kdeplot(df_scale['Salary'], ax=ax2)

plt.show()
```
